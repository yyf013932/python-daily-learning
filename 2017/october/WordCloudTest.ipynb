## WordCloud库使用 github:https://github.com/amueller/word_cloud

API Reference https://amueller.github.io/word_cloud/references.html

```python
#coding:utf-8 
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
#这里是为了在程序里显示图片
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
#如果需要显示中文 必须设定能显示中文的字体
font = 'C:\Windows\Fonts\simfang.ttf'
wc = WordCloud(font_path=font,width=800,height=600,min_font_size =6,max_words=100)
```

```python
f = open('text.txt',encoding='utf-8')
text = f.read()
f.close()
```

```python
wc.generate(text)
```

```python
plt.figure()  
# 以下代码显示图片  
plt.imshow(wc)  
plt.axis("off")  
plt.show() 
```

```python
wc.to_file('out_no_fenci.png')
```

### 之前的未进行分词，所以结果并不好，下面进入分词

```python
import jieba
```

```python
#进行分词
wordList =jieba.cut(text)
#去停止词
stopwords=['的','了','呢','着','罢','\n','，','？','。','；','：','—']
textNew = ' '.join([w for w in wordList if w not in stopwords])
```

```python
wc = WordCloud(font_path=font,width=800,height=600,min_font_size =6,max_words=100)
wc.generate(textNew)

plt.figure()  
# 以下代码显示图片  
plt.imshow(wc)  
plt.axis("off")  
plt.show() 

wc.to_file('out_fenci.png')
```

## 还可以根据图片生成相应形状的词云图(mask)

```python
from scipy.misc import imread
back_coloring = imread("mask.png")  
```

```python
wc = WordCloud(font_path=font,width=800,height=600,min_font_size =6,max_words=100,
                        mask=back_coloring)
wc.generate(textNew)

plt.figure()  
# 以下代码显示图片  
plt.imshow(wc)  
plt.axis("off")  
plt.show() 

wc.to_file('out_fenci_back.png')
```
